## Pandas Built-in Data Visualization

```python
import numpy as np
import seaborn as sns
import pandas as pd
df1 = pd.read_csv('df1', index_col=0)
df2 = pd.read_csv('df2')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist', bins=30)
```

```python
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.8)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.head()
```

```python
df1['B'].plot.line(figsize=(12,3), lw=0.8)
```

```python
df1.plot.scatter(x='A', y='B')
```

```python
df1.plot.scatter(x='A', y='B', c='C')
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100)
```

```python
df2.plot.box()
```
