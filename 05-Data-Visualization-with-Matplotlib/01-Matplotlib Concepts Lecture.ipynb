## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
# FUNCTIONAL
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

### Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1]) # shape

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.5,0.4])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGE PLOT')
axes2.set_title('SMALL PLOT')
```
