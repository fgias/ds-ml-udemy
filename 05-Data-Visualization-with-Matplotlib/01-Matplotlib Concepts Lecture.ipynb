## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
# FUNCTIONAL
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

### Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1]) # shape

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.5,0.4])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGE PLOT')
axes2.set_title('SMALL PLOT')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout() # if ticks overlap with graphs
```

```python
fig = plt.figure(figsize=(3,2)) # size is in inches
# dpi, argument `dpi=...`

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8,4))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('test.png', dpi=50)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.legend(loc=0) # default is loc = 0: "best position"
```
